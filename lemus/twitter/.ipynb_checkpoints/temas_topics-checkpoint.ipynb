```python
import re
import pandas as pd
from nltk.corpus import stopwords
```

```python
df_coments = pd.read_csv("coments_x.csv")
df_coments = df_coments.dropna()
df_coments
```

```python
def limpiar_texto(texto):
    # Eliminar caracteres especiales y números
    texto_limpio = re.sub(r'[^\w\sáéíóúüÁÉÍÓÚÜñ]', '', texto, flags=re.UNICODE)
    # Eliminar espacios en blanco adicionales y convertir a minúsculas
    texto_limpio = ' '.join(texto_limpio.split()).lower()
    return texto_limpio

df_coments["texto"] = df_coments["tokens"].apply(limpiar_texto)
df_coments_clean = df_coments.copy()
df_coments_clean
```

```python
##########NUBE DE PALABRAS
def preprocess_text(text):
    words = text # Tokenizar y convertir a minúsculas
    words = [word for word in words if word.isalpha()]  # Eliminar caracteres no alfabéticos
    words = [word for word in words if word not in stopwords]  # Eliminar palabras vacías
    return words


preprocessed_text = df_coments_clean["tokens"].apply(preprocess_text)

preprocessed_text
```
